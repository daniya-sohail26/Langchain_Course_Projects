### Tavily API Setup

To use Tavily search, you'll need an API key. Please enter it when prompted.

```python
!pip install langchain-community
from langchain_community.tools.tavily_search import TavilySearchResults
import getpass
import os

if not os.getenv("TAVILY_API_KEY"):
    os.environ["TAVILY_API_KEY"] = getpass.getpass("Enter your Tavily API key: ")
```

```python
tavily_tool = TavilySearchResults(max_results=5)
```

```python
import getpass
import os

if not os.getenv("OPENROUTER_API_KEY"):
    os.environ["OPENROUTER_API_KEY"] = getpass.getpass("Enter your OpenRouter API key: ")
```

The Tavily Search tool primarily takes a text query. To use the multimodal `HumanMessage` (`multimodal_question`) as context, we will extract its text component for the Tavily query. The image content (`img_b64`) cannot be directly processed by the Tavily search tool itself, but a sophisticated agent could interpret it and formulate a more precise text query.

```python
from langchain_openrouter import ChatOpenRouter

model = ChatOpenRouter(
    model="openrouter/free",
    temperature=0,
    max_tokens=1024,
    max_retries=2,
)
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

agent=create_agent(model=model,
                   checkpointer=InMemorySaver(),
                   )
```

```python
from ipywidgets import FileUpload
from IPython.display import display

uploader = FileUpload(accept='.png', multiple=False)
display(uploader)
```

```python
print(uploader.value)
```

```python
import base64

uploaded_files_dict = uploader.value # uploader.value is a dictionary like {filename: {metadata: ..., content: ...}}

if uploaded_files_dict:
    # Get the dictionary corresponding to the uploaded file (assuming only one file due to multiple=False)
    file_data = list(uploaded_files_dict.values())[0]

    content_mv = file_data["content"]

    img_byte=bytes(content_mv)

    img_b64= base64.b64encode(img_byte).decode("utf-8")
else:
    print("No file was uploaded.")
    img_b64 = None # Handle case where no file is uploaded
```

```python
from langchain.messages import HumanMessage

multimodal_question=HumanMessage(
    content=[
        {"type":"text", "text":"What dishes can I make with the ingredients you can see in my fridge?"},
        {"type":"image", "base64":img_b64,"mime_type":"image/png"}
    ]
)
```

```python
# Extract the text part from the multimodal question for the Tavily query
original_tavily_query = ""
for part in multimodal_question.content:
    if isinstance(part, dict) and part.get("type") == "text":
        original_tavily_query = part.get("text")
        break

# Get the agent's response which should include image interpretation
agent_response = agent.invoke(
    {"messages": [multimodal_question]},
    config={"configurable": {"thread_id": "default_thread"}}
)

# Combine original text query with agent's interpretation for Tavily
combined_tavily_query = original_tavily_query
if hasattr(agent_response, 'content') and agent_response.content:
    # Assuming the agent's content is a string that can be appended or used directly
    combined_tavily_query = f"{original_tavily_query} based on ingredients: {agent_response.content}"


if combined_tavily_query:
    print(f"Using Tavily with combined query: '{combined_tavily_query}'")
    tavily_result = tavily_tool.invoke({"query": combined_tavily_query})
    from pprint import pprint
    pprint(tavily_result)
else:
    print("No text found or generated to use for Tavily search.")
```

```python
from langchain.messages import HumanMessage
import pprint

# Create a new message with the Tavily search results
tavily_results_message = HumanMessage(
    content=f"Here are the search results from Tavily based on your fridge contents: {pprint.pformat(tavily_result)}\n\nNow, considering these results and the ingredients you asked about earlier, what dishes can you recommend?"
)

response = agent.invoke(
    {"messages": [multimodal_question, tavily_results_message]},
    config={"configurable": {"thread_id": "default_thread"}}
)
```

```python
from pprint import pprint
pprint(response["messages"][1].content)
```

### Notebook Widget Metadata Repair Script

The script below will help fix the 'Invalid Notebook' error by cleaning corrupted ipywidgets metadata from your notebook file. **Please read the instructions above and run this script in a *new*, separate Colab notebook.**

```python
import nbformat
import os

def fix_notebook_widgets_metadata(notebook_path):
    """
    Removes the 'widgets' metadata section from a Jupyter notebook file
    to fix issues related to corrupted ipywidgets state.
    """
    try:
        with open(notebook_path, 'r', encoding='utf-8') as f:
            nb = nbformat.read(f, as_version=4)

        if 'widgets' in nb.metadata:
            del nb.metadata['widgets']
            print(f"Successfully removed 'widgets' metadata from {notebook_path}")
        else:
            print(f"'widgets' metadata not found in {notebook_path}. No changes made.")

        with open(notebook_path, 'w', encoding='utf-8') as f:
            nbformat.write(nb, f)
        print("Notebook saved with cleaned metadata.")

    except FileNotFoundError:
        print(f"Error: Notebook file not found at '{notebook_path}'")
    except Exception as e:
        print(f"An error occurred while processing the notebook: {e}")

# --- IMPORTANT: Replace 'your_corrupted_notebook.ipynb' with your actual notebook file path ---
# You will likely need to mount Google Drive first if your file is there.
# Example for mounting Drive:
# from google.colab import drive
# drive.mount('/content/drive')
# Then, use a path like: '/content/drive/MyDrive/YourNotebookName.ipynb'

notebook_file_to_fix = 'your_corrupted_notebook.ipynb' # <--- CHANGE THIS LINE IN THE *NEW* NOTEBOOK

if notebook_file_to_fix == 'your_corrupted_notebook.ipynb':
    print("Please update the 'notebook_file_to_fix' variable with the actual path to your notebook in the *new* Colab notebook.")
else:
    fix_notebook_widgets_metadata(notebook_file_to_fix)
```
